# Level 2: MC Cafe

our instructors love coffee so much that they decided to open a cafe! Of course, we will need the help of our students to develop the ordering app that the customers can use.

On the cafe menu, there are three categories for the items we sell - coffee, tea, and pastries.

* Coffee
     - Espresso = 1.50
     - Americano - 2.00
     - Caffe latte - 2.25
     - Mocha - 2.25
     - Cappucino - 2.50
     - Macchiato - 2.50
* Tea
     - White tea - 1.50
     - Green tea - 2.00
     - Oolong tea - 2.25
     - Black tea - 2.25
     - Earl Grey - 2.50
     - Matcha - 2.50
* Pastries
     - Croissant - 5.50
     - Cinnamon roll - 4.50
     - Banana bread - 4.25
     - Pumpkin bread - 3.75
     - Sticky bun - 4.25
     - Apple turnover - 3.75

When your program first starts, you must greet the customers by saying, "Welcome to MC Cafe!" Then ask the customer if they would like to see the menu. If yes, then display the menu items and prices.

Then ask the user if they would like to order. If yes, then prompt the customer to enter an item and the quantity that they want. After they input an item, then ask them if they would like to order another item.

Once your customers are finished with the order input, then display an order receipt that lists the items that they ordered and the quantity. Then at the bottom of the receipt, display the subtotal (the amount of the item before tax), the amount in sales tax (our business is in Maryland, so state tax is 6%), and finally their grand total (subtotal plus sales tax).

At the end of their transaction, display a message to our customers saying, "Thank you for visiting MC Cafe!"

**BONUS:** Add a feature that tells a customer when an item they attempted to order is not on the menu.

```python
def show_menu (m):
    for category, sub_menu in m.items():
        print(category)
        for item, price in sub_menu.items():
            print(f"\t{item} - ${price:.2f}")

def make_order (sub_menu, receipt):
    user_item = input("Please select item: ")
    quantity = int(input("How many would you like?: "))

    if user_item in receipt:
        receipt[user_item] += quantity
    else :
        receipt[user_item] = quantity
    
    return sub_menu[user_item] * quantity
    
            
coffee = {"Espresso":1.50, "Americano":2.00, "Caffe latte":2.25, "Mocha":2.25, "Cappucino":2.50, "Macchiato":2.50}
tea = {"White tea":1.50, "Green tea":2.00, "Oolong tea":2.25, "Black tea":2.25, "Earl Grey":2.50, "Matcha":2.50}
pastries = {"Croissant":5.50, "Cinnamon roll":4.50, "Banana bread":4.25, "Pumpkin bread":3.75, "Sticky bun":4.25, "Apple turnover":3.75}

menu = {"Coffee":coffee, "Tea":tea, "Pastries":pastries}

receipt = {}
sub_total = 0

print("Welcome to MC Cafe!")

choice = input("Would you like to see the menu? (y/n): ")
choice = choice.lower()

if choice == 'y' or choice == 'yes':
    show_menu(menu)

choice = input("Would you like to make an order? (y/n): ")
choice = choice.lower()

while choice == 'y' or choice == 'yes':
    print("What would you like?")
    category = input("Please choose Coffee, Tea or Pastries: ")
    category = category.lower()
    
    if category == 'coffee' :
        sub_total += make_order(coffee,receipt)
                     
    elif category == 'tea' :
        sub_total += make_order(tea,receipt) 
    
    elif category == 'pastries' or category == 'pastry' :
        sub_total += make_order(pastries,receipt)
        
    else :
        print("Try again.")
        continue
    
    choice = input("Will there be anything else? (y/n): ")
    
print("\nYour Order:")
for item, quantity in receipt.items():
    print(f"{item} - {quantity}")

print(f"\nSub-Total:\t ${sub_total:.2f}")
print(f"Sales Tax:\t ${(sub_total*0.06):.2f}")
print(f"Total:\t\t ${(sub_total*1.06):.2f}")

print("\nThank you for visiting MC Cafe!")
```
